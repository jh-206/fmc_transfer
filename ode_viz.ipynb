## Show demonstrated equilibria timelag

```python
# NOTE, exact timelag numbers 1, 10, 100 don't plot nice
# making a representative plot and removing x axis

E = 20
m0 = 1
tsteps = 1500
def tlag(T=10):
    mt = np.zeros(tsteps)
    mt[0] = m0
    for t in np.arange(1, tsteps):
        mt[t] = (1-np.exp(-1/T)) * E + np.exp(-1/T)*mt[t-1]

    return mt

m1    = tlag(T=10)
m10   = tlag(T=50)
m100  = tlag(T=200)
m1000  = tlag(T=400)

plt.plot(m1, label="FM1", color=plot_styles["fm1"]["color"], linewidth=2.5)
plt.plot(m10, label="FM10", color=plot_styles["fm"]["color"], linewidth=2.5)
plt.plot(m100, label="FM100", color=plot_styles["fm100"]["color"], linewidth=2.5)
plt.plot(m1000, label="FM1000", color=plot_styles["fm1000"]["color"], linewidth=2.5)

plt.axhline(y=20, color='k', linestyle='--', linewidth=2.5)
plt.text(1.0, 20.5, "Equilibrium", color='k', fontsize=14, ha='right', va='bottom',
         transform=plt.gca().get_yaxis_transform())

plt.ylim(0, 23)
plt.ylabel("FMC (%)", fontsize=14)
# plt.xlabel('Time', fontsize=14)
plt.xticks([])
plt.legend(title="Fuel Class", fontsize=13, title_fontsize=14)
plt.savefig("outputs/timelag_steady.png", dpi=300)
```

```python
E0 = 12          # mean equilibrium FMC
A = 10            # amplitude
P = 500          # heuristic period in timesteps
m0 = 12
tsteps = 1000

def equilibrium_input(t, mean=E0, amp=A, period=P):
    return mean + amp * np.sin(2 * np.pi * t / period)

def tlag(T=10):
    mt = np.zeros(tsteps)
    mt[0] = m0
    for t in np.arange(1, tsteps):
        Et = equilibrium_input(t)
        mt[t] = (1 - np.exp(-1 / T)) * Et + np.exp(-1 / T) * mt[t - 1]
    return mt

m1 = tlag(T=10)
m10 = tlag(T=50)
m100 = tlag(T=200)
m1000 = tlag(T=400)

plt.plot(m1, label="FM1", color=plot_styles["fm1"]["color"], linewidth=2.5)
plt.plot(m10, label="FM10", color=plot_styles["fm"]["color"], linewidth=2.5)
plt.plot(m100, label="FM100", color=plot_styles["fm100"]["color"], linewidth=2.5)
plt.plot(m1000, label="FM1000", color=plot_styles["fm1000"]["color"], linewidth=2.5)

plt.ylim(0, 23)
plt.ylabel("FMC (%)", fontsize=14)
plt.xticks([])
plt.xlabel("")
plt.legend(title="Fuel Class", fontsize=13, title_fontsize=14)
plt.savefig("outputs/timelag_wave.png", dpi=300)
```
