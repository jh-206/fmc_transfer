```python
def plot_bars(means, stds=None, ax=None, color="#1f77b4"):
    methods = [
        "No Transfer",
        "Full Fine-Tuning",
        "Freeze Recurrent Layer",
        "Freeze Dense/Output Layers",
        "Time-Warping",
        "Time-Warping + Fine-Tuning"
    ]

    if ax is None:
        fig, ax = plt.subplots(figsize=(6, 4), dpi=DPI)

    y = np.arange(len(methods))

    if stds is not None:
        # background (white, thicker)
        ax.barh(y, means, xerr=stds, capsize=4,
                ecolor="white", error_kw=dict(linewidth=3, capthick=3),
                color=color)
        
        # foreground (actual color, thinner)
        ax.barh(y, means, xerr=stds, capsize=4,
                ecolor="black", error_kw=dict(linewidth=1.5, capthick=1.5),
                color=color)

    ax.set_yticks(y)
    ax.set_yticklabels(methods, fontsize=TICK_SIZE)
    ax.set_xlabel("RSME (%)", fontsize=LABEL_SIZE)
    ax.tick_params(axis='x', labelsize=TICK_SIZE)

    ax.invert_yaxis()

    return ax
```

```python
# document-safe plot defaults
DPI = 300
LABEL_SIZE = 14
TICK_SIZE = 12
CBAR_LABEL_SIZE = 13
```

## FM1

```python
ax = plot_bars(
    means=[
        notl["FM1"]["Mean Value"][0],
        full["FM1"]["Mean Value"][0],
        frzr["FM1"]["Mean Value"][0],
        frzd["FM1"]["Mean Value"][0],
        tw["FM1"]["Mean Value"][0],
        twf["FM1"]["Mean Value"][0],
    ],
    stds=[
        notl["FM1"]["Std"][0],
        full["FM1"]["Std"][0],
        frzr["FM1"]["Std"][0],
        frzd["FM1"]["Std"][0],
        tw["FM1"]["Std"][0],
        twf["FM1"]["Std"][0],
    ]
)

for bar in ax.patches:
    bar.set_color(plot_styles["fm1"]["color"])

ax.set_title("FM1 RMSE Comparison", fontsize=LABEL_SIZE)
plt.tight_layout()
plt.savefig("../outputs/fm1_rmse.png")
```
